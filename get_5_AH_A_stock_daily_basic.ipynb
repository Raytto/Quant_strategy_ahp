```python
# 导入tushare
import tushare as ts
import pandas as pd
import my_settings
```

```python
ahldf = pd.read_csv(r"data\AH_stock_list.csv", encoding="utf-8-sig")
ahldf
```

```python
# 分批次拉 A 股市值等基本数据

pro = ts.pro_api(my_settings.tushare_api)
result = None
for idx, a_ts_code in ahldf["ts_code_A"].items():
    print(a_ts_code)
    # 初始化pro接口

    # 拉取数据
    df = pro.daily_basic(
        **{
            "ts_code": a_ts_code,
            "trade_date": "",
            "start_date": my_settings.date_start,
            "end_date": my_settings.date_end,
            "limit": "",
            "offset": "",
        },
        fields=[
            "ts_code",
            "trade_date",
            "close",
            "turnover_rate",
            "turnover_rate_f",
            "volume_ratio",
            "pe",
            "pe_ttm",
            "pb",
            "ps",
            "ps_ttm",
            "dv_ratio",
            "dv_ttm",
            "total_share",
            "float_share",
            "free_share",
            "total_mv",
            "circ_mv",
        ]
    )
    if result is None:
        result = df
    else:
        result = pd.concat([result, df], ignore_index=True)
# save to csv
result.to_csv("data\AH_stock_daily_A_basic.csv", index=False, encoding="utf-8-sig")
```
