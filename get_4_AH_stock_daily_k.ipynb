```python
import pandas as pd

# 获取AH的list



ahdf = pd.read_csv("data\AH_stock_list.csv", encoding="utf-8-sig")
```

# 分批次拉 A 股 K线数据

```python
# 分批次拉 A 股 K线数据
# 导入tushare
import tushare as ts
import my_settings

# 初始化pro接口
pro = ts.pro_api(my_settings.tushare_api)

result = None
for idx, a_ts_code in ahdf["ts_code_A"].items():
    print(f"{idx} , {a_ts_code}")
    # 拉取数据
    df = pro.daily(
        **{
            "ts_code": a_ts_code,
            "trade_date": "",
            "start_date": my_settings.date_start,
            "end_date": my_settings.date_end,
            "limit": "",
            "offset": "",
        },
        fields=[
            "ts_code",
            "trade_date",
            "open",
            "high",
            "low",
            "close",
            "pre_close",
            "amount",
        ],
    )
    if result is None:
        result = df
    else:
        result = pd.concat([result, df], ignore_index=True)
```

```python
# save to csv
result.to_csv("data\AH_stock_daily_A_k.csv", index=False, encoding="utf-8-sig")
```

# 分批次拉 A 股复权因子数据

```python
# 导入tushare
import tushare as ts

# 初始化pro接口
pro = ts.pro_api(my_settings.tushare_api)

result = None
for idx, a_ts_code in ahdf["ts_code_A"].items():
    print(f"{idx} , {a_ts_code}")
    # 拉取数据
    df = pro.adj_factor(
        **{
            "ts_code": a_ts_code,
            "trade_date": "",
            "start_date": my_settings.date_start,
            "end_date": my_settings.date_end,
            "limit": "",
            "offset": "",
        },
        fields=["ts_code", "trade_date", "adj_factor"],
    )
    if result is None:
        result = df
    else:
        result = pd.concat([result, df], ignore_index=True)
# save to csv
result.to_csv("data\AH_stock_daily_A_adj_factor.csv", index=False, encoding="utf-8-sig")
```

# 分批次拉 HK 股 K线和复权数据

```python
# 导入tushare
import tushare as ts
import time

# 初始化pro接口
pro = ts.pro_api(my_settings.tushare_api)

result = None
count = 0
for idx, a_ts_code in ahdf["ts_code_HK"].items():
    print(f"K {idx} , {a_ts_code}")
    # 拉取数据
    df = pro.hk_daily(
        **{
            "ts_code": a_ts_code,
            "trade_date": "",
            "start_date": my_settings.date_start,
            "end_date": my_settings.date_end,
            "limit": "",
            "offset": "",
        },
        fields=[
            "ts_code",
            "trade_date",
            "open",
            "high",
            "low",
            "close",
            "pre_close",
            "amount",
        ],
    )
    if result is None:
        result = df
    else:
        result = pd.concat([result, df], ignore_index=True)
    count += 1
    if count == 1:
        # time.sleep(60)
        count = 0
# save to csv
result.to_csv("data\AH_stock_daily_HK_k.csv", index=False, encoding="utf-8-sig")
```

# 分批次拉 HK 股 复权信息数据

```python
result = None
count = 0
for idx, a_ts_code in ahdf["ts_code_HK"].items():
    print(f"adj_factor {idx} , {a_ts_code}")
    # 拉取数据
    df = pro.hk_daily_adj(
        **{
            "ts_code": a_ts_code,
            "trade_date": "",
            "start_date": my_settings.date_start,
            "end_date": my_settings.date_end,
            "offset": "",
            "limit": "",
            "symbol": "",
        },
        fields=[
            "ts_code",
            "trade_date",
            "close",
            "open",
            "high",
            "low",
            "pre_close",
            "change",
            "pct_change",
            "vol",
            "amount",
            "vwap",
            "adj_factor",
            "turnover_ratio",
            "free_share",
            "total_share",
            "free_mv",
            "total_mv",
        ],
    )
    if result is None:
        result = df
    else:
        result = pd.concat([result, df], ignore_index=True)
    count += 1
    if count >= 1:
        # time.sleep(60)
        count = 0
# save to csv
result.to_csv(
    "data\AH_stock_daily_HK_adj_factor.csv", index=False, encoding="utf-8-sig"
)
```
