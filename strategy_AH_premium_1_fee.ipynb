我有如图的数据表，其中：
* trade_date : 日期
* p_a_adj : 股票对应日期的A股的复权前复权价格
* p_hk_adj : 股票对应日期的H股的复权前复权价格
* hk2cn : 当日的港币兑人民币汇率（1港币可以换多少人民币）
* AH_premium_adj : A股相对于H股当日的溢价

我希望构建一个基于AH溢价的选股策略：
* 首先帮我写一个函数：
    * 其参数是日期
    * 输出是两个 list ，一个list是AH溢价最高的5支股票名字，一个是AH溢价最低的5支股票名字
* 以及一个交易策略回测函数：
    * 待定参数是策略周期:X个月
    * 每个周期一开始把手上的钱均分成10份，其中5份对应平均购买那5支AH溢价最高的股票的H股，另外5份平均购买AH溢价最低的股票的A股
    * 每个周期结束后进入下一个周期，用同样策略重新配置仓位
    * 考虑资金量和交易成本（H股交易双向都按0.05%计算，A股交易双向都按0.01%计算，几个值都用变量来配置）
* 分别绘制出策略周期X为1个月、2个月、3个月、6个月情况下的策略收益曲线，从2018年开始模拟



```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# ———— 配置中文字体 & 负号 ————
matplotlib.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei"]
matplotlib.rcParams["axes.unicode_minus"] = False


def get_top_bottom(
    ah_df: pd.DataFrame, date: str | pd.Timestamp, top_n: int = 5
) -> tuple[list[str], list[str]]:
    """返回指定交易日 date 当天 AH 溢价最高和最低的股票名称列表。"""
    df = ah_df.copy()
    # 转 datetime
    if df["trade_date"].dtype == object or np.issubdtype(
        df["trade_date"].dtype, np.number
    ):
        df["trade_date"] = pd.to_datetime(df["trade_date"].astype(str), format="%Y%m%d")
    if isinstance(date, str):
        date = pd.to_datetime(date, format="%Y%m%d")
    today = df[df["trade_date"] == date].dropna(subset=["AH_premium_adj"])
    top5 = today.nlargest(top_n, "AH_premium_adj")["name"].tolist()
    bottom5 = today.nsmallest(top_n, "AH_premium_adj")["name"].tolist()
    return top5, bottom5
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# ———— 配置中文字体 & 负号 ————
matplotlib.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei"]
matplotlib.rcParams["axes.unicode_minus"] = False


def get_top_bottom(
    ah_df: pd.DataFrame, date: str | pd.Timestamp, top_n: int = 5
) -> tuple[list[str], list[str]]:
    """返回指定交易日 date 当天 AH 溢价最高和最低的股票名称列表。"""
    df = ah_df.copy()
    # 转 datetime
    if df["trade_date"].dtype == object or np.issubdtype(
        df["trade_date"].dtype, np.number
    ):
        df["trade_date"] = pd.to_datetime(df["trade_date"].astype(str), format="%Y%m%d")
    if isinstance(date, str):
        date = pd.to_datetime(date, format="%Y%m%d")
    today = df[df["trade_date"] == date].dropna(subset=["AH_premium_adj"])
    top5 = today.nlargest(top_n, "AH_premium_adj")["name"].tolist()
    bottom5 = today.nsmallest(top_n, "AH_premium_adj")["name"].tolist()
    return top5, bottom5
```

```python
def backtest_ah_strategy(
    ah_df: pd.DataFrame,
    start_date: str = "20180101",
    months: int = 1,
    initial_capital: float = 1.0,
) -> pd.DataFrame:
    """
    AH 溢价轮动回测（离散周期估值），返回每期末的净值曲线。
    """
    df = ah_df.copy()
    df["trade_date"] = pd.to_datetime(df["trade_date"].astype(str), format="%Y%m%d")
    df = df.sort_values("trade_date")
    df["ph_cny"] = df["p_hk_adj"] * df["hk2cn"]

    # 所有可交易日
    dates = df["trade_date"].drop_duplicates().sort_values()
    dates = dates[dates >= pd.to_datetime(start_date, format="%Y%m%d")]
    if dates.empty:
        raise ValueError("start_date 之后没有数据")

    # 构造周期起始/结束列表
    period_starts = []
    cur = dates.min()
    last = dates.max()
    while cur <= last:
        ps = dates[dates >= cur].min()
        period_starts.append(ps)
        cur = ps + pd.DateOffset(months=months)
    period_starts = [d for d in period_starts if d <= last]

    period_ends = []
    for i in range(len(period_starts) - 1):
        nxt = period_starts[i + 1]
        pe = dates[dates < nxt].max()
        period_ends.append(pe)
    period_ends.append(last)

    capital = initial_capital
    records = []
    for ps, pe in zip(period_starts, period_ends):
        # 选股
        df_ps = df[df["trade_date"] == ps]
        top5, bot5 = get_top_bottom(df_ps, ps, top_n=5)
        alloc = capital / 10

        # 买入份额
        buy_ps = df_ps.set_index("name")
        shares_top = alloc / buy_ps.loc[top5, "ph_cny"]
        shares_bot = alloc / buy_ps.loc[bot5, "p_a_adj"]

        # 平仓：取 <= pe 时每只股票最后一条价格
        df_end = (
            df[df["trade_date"] <= pe]
            .sort_values("trade_date")
            .groupby("name", as_index=True)
            .last()
        )
        p_top = df_end["ph_cny"].reindex(top5).fillna(0)
        p_bot = df_end["p_a_adj"].reindex(bot5).fillna(0)

        val_top = (shares_top.values * p_top.values).sum()
        val_bot = (shares_bot.values * p_bot.values).sum()
        capital = val_top + val_bot

        records.append({"date": pe, "equity": capital})

    return pd.DataFrame(records)
```

```python
if __name__ == "__main__":
    # —— 0. 设置 —— #
    start_dt_str = "20180101"
    start_dt = pd.to_datetime(start_dt_str, format="%Y%m%d")
    initial_capital = 1_000_000
    # —— 1. 读数据 —— #
    ah_k = pd.read_csv("data/AH_premium_analysis.csv", encoding="utf-8-sig")
    SH300_k = pd.read_csv("data/SH300_daily_k.csv", encoding="utf-8-sig")
    HSI_k = pd.read_csv("data/HSI_daily_k.csv", encoding="utf-8-sig")

    # —— 2. 准备指数归一化序列 —— #
    for idx_df in (SH300_k, HSI_k):
        idx_df["trade_date"] = pd.to_datetime(
            idx_df["trade_date"].astype(str), format="%Y%m%d"
        )
    sh300 = (
        SH300_k[SH300_k["trade_date"] >= start_dt]
        .sort_values("trade_date")
        .set_index("trade_date")
    )
    hsi = (
        HSI_k[HSI_k["trade_date"] >= start_dt]
        .sort_values("trade_date")
        .set_index("trade_date")
    )
    sh300["norm"] = sh300["close"] / sh300["close"].iloc[0]
    hsi["norm"] = hsi["close"] / hsi["close"].iloc[0]

    # —— 3. 回测并绘图 —— #
    plt.figure(figsize=(12, 6))

    # 3.1 画指数
    plt.plot(sh300.index, sh300["norm"], label="沪深300 指数", linewidth=1.2)
    plt.plot(hsi.index, hsi["norm"], label="恒生指数", linewidth=1.2)

    # 3.2 画策略（1、2、3、6 个月）
    for m in [1, 2, 3, 6]:
        eq = backtest_ah_strategy(ah_k, start_date=start_dt_str, months=m)
        plt.plot(
            eq["date"],
            eq["equity"],
            label=f"{m} 个月轮动策略",
            linestyle="--",
            marker="o",
        )

    plt.title(f"AH 溢价策略 vs. 两大指数（{start_dt_str} 起归一化）")
    plt.xlabel("日期")
    plt.ylabel("归一化收益（基准=1）")
    plt.legend(loc="upper left", fontsize=9)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
